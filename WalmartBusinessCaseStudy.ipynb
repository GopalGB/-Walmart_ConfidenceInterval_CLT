```python
df_male_customer = df_cust_gender.loc[df_cust_gender['Gender'] == 'M'] 
df_female_customer = df_cust_gender.loc[df_cust_gender['Gender'] == 'F']

sample_size_trend = [] 

# Calculate the maximum sample size based on available data
max_sample_size = len(df_male_customer['Purchase'])

for i in range(500, max_sample_size):
    sample = df_male_customer['Purchase'].sample(i)
    sample_mean = sample.mean()
    sample_size_trend.append(sample_mean) 

    # Debugging: Print sample size and mean
    print(f"Sample Size: {i}, Sample Mean: {sample_mean}")

plt.plot(sample_size_trend) 
plt.xlabel("sample_size") 
plt.ylabel("sample_mean")
plt.show()
```
